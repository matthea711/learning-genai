# OpenAI Embeddings

Note: Embedding -- mathematical representation of data

## Creating Embeddings (Basics)

```python
from openai import OpenAI

# Creating the client
client = OpenAI()

# Text for embedding
my_text = 'red'
my_text = my_text.replace('\n', '')   # Basic cleanup for text used for embedding purposes

# Creating Embedding
response = client.embeddings.create(
  input=my_text,
  model='text-embedding-3-small'
)

# print(response.data[0].embedding)
my_print(response.data[0].embedding)
```

## Embedding Method, Panda Dataframes & Generating CSV File

```python
import pandas as pd
from openai import OpenAI
# OpenAI assholes removed the following helpful utils
# from openai.embeddings_utils import get_embedding, cosine_similarity [disappeared after openai v1.0]

# Initialize the OpenAI client
client = OpenAI()

# Function to get embeddings for a given text
def get_embedding(text, model="text-embedding-3-small"):
  text = text.replace("\n", " ")
  return client.embeddings.create(input=[text], model=model).data[0].embedding

# Create a simple DataFrame w/ pandas
data = {'my_text': ['This is a test.', 'Another test sentence. ']}
df = pd.DataFrame(data)

# Apply the get_embedding function to each row in the 'combined' column
df ['openai_embedding'] = df.my_text.apply(lambda x: get_embedding(x, model='text-embedding-3-small' ))

# Save the DataFrame with embeddings to a CSV file
df.to_csv('embedded_1k_moose.csv', index=False)
```

```python
# Reading the data from the csv file
df = pd.read_csv('embedded_1k_moose.csv')
print(df)
```

## Calculating Tokens

```python
import tiktoken

# Calculating number of tokens in data set
enc = tiktoken.encoding_for_model('text-embedding-3-small')

print('Showing Token ID-s for each word...')
print(enc.encode ('above us only sky'))
print('Showing Total number of Tokens for the sentence...')
print(len(enc.encode('above us only sky')))
```

## Calculating Embedding Cost (Ball Park Idea)

```python
words = list(df['my_text'])           # Getting my_text column in dataframe
my_print(words)
# print(words)

print('-' * 100)

# Calculating total tokens
total_tokens = sum([len(enc.encode(word)) for word in words])
print( 'Total Tokens: ', total_tokens)

print('-' * 100)

# Calculating the cost
cost = total_tokens * (0.0004/1000)
print(f'Estimated cost in USD: ${cost: 10f}')
```
